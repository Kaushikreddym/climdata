```python
import json
from climdata.extremes.indices import extreme_index 
geojson = {
  "type": "FeatureCollection",
  "features": [
    {
      "type": "Feature",
      "properties": {},
      "geometry": {
        "coordinates": [
          52.246667038198012,
          16.891982026993958
        ],
        "type": "Point"
      }
    }
  ]
}


import climdata
clim_cfg, clim_ds = climdata.extract_data(
    overrides=[
            "dataset=cmip",
            "variables=['pr']",
            f"time_range.start_date=1989-01-01",
            f"time_range.end_date=2014-12-31",
            f"aoi='{json.dumps(geojson)}'",
            "dsinfo.mswx.params.google_service_account=./.climdata_conf/service.json",
    ]
,
save_to_file=False
)

indices = extreme_index(clim_cfg, clim_ds)
index = indices.calculate('SPI1').compute()
index = index.compute()
```
